```python
#!/usr/bin/env python
""" Access data in the shared memory used by rtcads.
"""
import os
import struct
import sysv_ipc
import re
import sys
```

```python
TMData_size = 128
TMData_fmt = "5f3I24s24s24s24s"

# Nul terminated C string
cstre = re.compile(b'([^\x00]*)\x00')
cstrmo = cstre.match(cstrval)
```

```python
# tm_file_list = os.listdir('./juda_byteshift_memo_data/')
tm_file_list = os.listdir('./ftm_telemetry_files/')
tm_file_list.sort()
testfile = './ftm_telemetry_files/' + tm_file_list[0]
```

```python
# GEN slots: MF, mf, fmt, Spare
GEN_base = 0
GEN_slots = 4
GEN_count = 3

# EVT slots: evt_MF, evt_mf, PSDspStatus, SSDspStatus, EEDspStatus, RD5DspStatus
EVT_base = GEN_base + GEN_slots
EVT_slots = 6
EVT_count = 6

# Secondary science data (SSD).
SSD_base = EVT_base + EVT_slots
SSD_slots = 128
SSD_count = 103

# Engineering data (EED)
EED_base = SSD_base + SSD_slots
EED_slots = 128
EED_count = 48

# These are followed by the image data.
```

```python
def extract_string(index):
    """Get the label, value string and conval for one parameter (ie TMData) 
    from DPP shared memory.
    index = item ordinal
    """
    # Reads up to byte_count bytes from the shared memory segment starting at offset and returns them as a bytes object (which is the same as a str under Python 2).

    print(f"INDEX IS {index}")
    
    with open(testfile, "rb") as f:
        full_ftm_binary = bytes(f.read())
    
    offset = (index * 128) + 128
    endpoint = ((index + 1) * 128) + 128
    
    tmdata = full_ftm_binary[offset:endpoint]
    
    # Extract one TMData.  Most entries here are undefined
    (yll, yul, rll, rul, conval, dnval, 
     flag, tag, cstrval, clabel, units, comment) = struct.unpack(TMData_fmt, tmdata)
    
#     print("CSTRVAL IS:", cstrval)
    
#     # Nul terminated C string
#     cstre = re.compile(b'([^\x00]*)\x00')
#     print(cstre)
#     cstrmo = cstre.match(cstrval)
    
#     print('CSTRE is:', cstre)
    
#     # Convert C strings
#     cstrmo = cstre.match(cstrval)
#     print(cstrmo)
#     if not cstrmo:
#         print('Formatted string does not match', cstrval)
#         sys.exit(1)
#     strval = cstrmo.group(1)
#     label = clabel.rstrip(b'\x00')
#     if not strval:
#         # Convert items with no strval to a string
#         strval = str(dnval)
    return (label, (strval, conval))
```

```python
def readSHMdata():
    """Extract strval and conval for all data items in DPP shared
    memory.
    """
    return dict([extract_string(i) for i in
                 list(range(GEN_base, GEN_base + GEN_count))
                 + list(range(EVT_base, EVT_base + EVT_count))
                 + list(range(SSD_base, SSD_base + SSD_count))
                 + list(range(EED_base, EED_base + EED_count))])
```

```python
data_dict = readSHMdata()
for x in data_dict:
    print(x, data_dict[x])
```
